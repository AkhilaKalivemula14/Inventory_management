```python
# importing libraries
import json
import time

# opening inventory file
fd=open('Inventory2.json','r')
js=fd.read()
fd.close()
record=json.loads(js)

# printing menu
print("------------------------------")
print('             Menu             ')
print("------------------------------")
for i in record.keys():
    
    print(i+' : '+record[i]['Name']+" | "+str(record[i]['Price'])+" | "+str(record[i]['Qt'])+"\n")

#printing user and purchase details
print("------------------------------")
ui_name=input("Enter name : ")
ui_phoneNumber=input("Enter phone number : ")
ui_mail=input("Enter mail ID : ")
ui_prod_id=input("Enter product ID : ")
ui_qt=input("Enter quantity : ")

# checking if required quantity is present
if int(ui_qt)<=record[ui_prod_id]['Qt']:
    print("------------------------------")
    print('Product Name   : ',record[ui_prod_id]['Name'])
    print('Price          : ',record[ui_prod_id]['Price'])
    print('Quantity       : ',ui_qt)
    print("------------------------------")
    print('             Bill             ')
    print("------------------------------")
    print('Billing Amount :',int(ui_qt)*record[ui_prod_id]['Price'])
    
    # recording customer and sale details
    sales = ui_name+","+ui_mail+","+ui_phoneNumber+","+ui_prod_id+","+record[ui_prod_id]["Name"]+","+str(ui_qt)+","+str(record[ui_prod_id]["Price"])+","+str(int(ui_qt) * record[ui_prod_id]["Price"])+","+time.ctime()+"\n"

    # updating inventory
    record[ui_prod_id]['Qt']=record[ui_prod_id]['Qt']-int(ui_qt)

# if the required quantity is not available
else:
    # asking user if he/she wants the available quantity
    print("------------------------------")
    print('only '+str(record[ui_prod_id]['Qt'])+' is available')
    print('do u want to proceed')
    ch=input('Press y to continue')
    if ch=='Y' or ch=='y':
        print("------------------------------")
        print('Product Name   : ',record[ui_prod_id]['Name'])
        print('Price          : ',record[ui_prod_id]['Price'])
        print('Quantity       : ',ui_qt)
        print("------------------------------")
        print('             Bill             ')
        print("------------------------------")
        print('Billing Amount :',int(ui_qt)*record[ui_prod_id]['Price'])
        
        # recoring customer and sale details
        sales = ui_name+","+ui_mail+","+ui_phoneNumber+","+ui_prod_id+","+record[ui_prod_id]["Name"]+","+str(record[ui_prod_id]['Qt'])+","+str(record[ui_prod_id]["Price"])+","+str(record[ui_prod_id]['Qt'] * record[ui_prod_id]["Price"])+","+time.ctime()+"\n"
        
        #updating inventory
        record[ui_prod_id]['Qt']=0
    else:
        print('Thanksyou for visiting')
        
js=json.dumps(record)

# Appending all sales details into Sales2.json file
fd=open('Sales2.json','a')
fd.write(sales)
fd.close()

# Updating inventory each time a purchase happens
fd=open('Inventory2.json','w')
fd.write(js)
fd.close()
print("------------------------------")
print('Inventory updated')
```
