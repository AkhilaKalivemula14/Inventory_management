```python
import time

fd=open('Inventory.txt','r')
products=fd.read().split('\n')
fd.close()

ui_username=input('Enter user name:')
ui_phone=input('Enter phone number:')
ui_mail=input('Enter mail ID:')
ui_prod_id=input("Enter product ID:")
ui_prod_qt=input("Enter product quantity:")
updated_list=[]
for product in products:
    product_details=product.split(',')
    
    if product_details[0]==ui_prod_id:
        if int(ui_prod_qt)<=int(product_details[3]):
            print('----------------------------')
            print('Product Name       :',product_details[1])
            print('Price              :',product_details[2])
            print('Quantity           :',ui_prod_qt)
            print('----------------------------')
            print('Billing Amount     :',str(int(product_details[2])*int(ui_prod_qt)))
            print('----------------------------')
            
            product_details[3]=str(int(product_details[3])-int(ui_prod_qt))
            
            fd = open("Sales.txt",'a')
            sales_detail = ui_username +","+ ui_phone +","+ ui_mail +","+product_details[1] +","+ ui_prod_id +","+ ui_prod_qt +","+ str(int(ui_prod_qt) * int(product_details[2]))+","+time.ctime()+ "\n"
            fd.write(sales_detail)
            fd.close()
             
        
        else :
            print('Sorry, the requested quantity is not available')
            print('The available quantity is '+product_details[3])
            print('If u want to continue press yes')
            ch=input('Press y/n:')
            if ch=='Y' or ch=='y':
                print('----------------------------')
                print('Product Name       :',product_details[1])
                print('Price              :',product_details[2])
                print('Quantity           :',product_details[3])
                print('----------------------------')
                print('Billing Amount     :',str(int(product_details[2])*int(product_details[3])))
                print('----------------------------')
                
                fd = open("Sales.txt",'a')
                sales_detail = ui_username +","+ ui_phone +","+ ui_mail +","+product_details[1] +","+ ui_prod_id +","+ product_details[3] +","+ str(int(product_details[3]) * int(product_details[2]))+","+time.ctime()+ "\n"
                fd.write(sales_detail)
                fd.close()
            
                product_details[3]='0'
            else:
                print('Thankyou')
        
    updated_list.append(product_details)

            
lst=[]
for i in updated_list:
    prod=i[0]+","+i[1]+","+i[2]+","+i[3]+"\n"
    lst.append(prod)
lst[-1]=lst[-1][:-1]

fd=open('Inventory.txt','w')
for i in lst:
    fd.write(i)
fd.close()
print('Inventory updated')
```
